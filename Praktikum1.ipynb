## Import Library

```python
import cv2
```

## Read Image

```python
img = cv2.imread('gambar_buah.png')
```

```python
img
```

## show image openCV

```python
cv2.imshow("gambar_buah.png",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## show image with Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

## RGB to Grayscale with OpenCV

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray, cmap='gray')
```

## Graycale to Binary with OpenCv

```python
(thresh, binary) = cv2.threshold(gray,128, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(binary, cmap='gray')
```
